Se construye la checklist con las cajas de verificaci√≥n y deacuerdo con la asignaci√≥n de tareas, la unidad

```python

df = pd.read_excel(ruta)

def mostrar_entrega(df_entrega, num_entrega):
    titulo = widgets.HTML(f"<h3>üì¶ Entrega {num_entrega}</h3>")
    items = []
    checkboxes = []

    def guardar(change):
        for cb, idx in checkboxes:
            df.loc[idx, 'Hecho'] = cb.value
        df.to_excel(ruta, index=False)

    for idx, row in df_entrega.iterrows():
        texto = f"""
        <b>Unidad {row['Unidad']} ‚Äì {row['Fase']}</b><br>
        üìù {row['Tarea']}<br>
        üë§ <i>{row['Responsable']}</i> | üóì {row['Fecha l√≠mite']}
        """
        cb = widgets.Checkbox(value=row['Hecho'])
        cb.observe(guardar, names='value')
        checkboxes.append((cb, idx))

        items.append(widgets.HBox([cb, widgets.HTML(texto)]))

    display(titulo, widgets.VBox(items))
```

```python

display(widgets.HTML("<h2>üìã Checklist CRISP-DM</h2>"))

for entrega in sorted(df['Entrega'].unique()):
    mostrar_entrega(df[df['Entrega'] == entrega], entrega)
```

## **2. Datos**
---

El dataset cuenta con un total aproximado de 20.700 registros, correspondientes a canciones, cada una descrita mediante 26 variables. Para facilitar su comprensi√≥n, las variables se agrupan en distintas categor√≠as seg√∫n el tipo de informaci√≥n que aportan sobre la canci√≥n. Las descripciones fueron extra√≠das directamente de la carta de datos del dataset en Kaggle.

La primera categor√≠a corresponde a las variables de identificaci√≥n, las cuales permiten reconocer y vincular cada canci√≥n dentro del conjunto de datos. Estas variables son de tipo categ√≥rico y no aportan directamente al an√°lisis t√©cnico del audio:

- `Track`: Nombre de la canci√≥n

- `Artist`: Artista de la canci√≥n

- `Url_spotify`: URL para acceder al contenido en Spotify

- `Uri`: C√≥digo utilizado para acceder a la canci√≥n mediante la API

- `Url_youtube`: URL para acceder a la canci√≥n en YouTube

- `Title`: T√≠tulo del video de la canci√≥n

- `Channel`: Canal de YouTube

- `Description`: Descripci√≥n del video en YouTube

- `Licensed`: Indica si el contenido est√° licenciado

- `official_video`: Variable booleana que indica si la URL de YouTube corresponde al video oficial

La siguiente categor√≠a corresponde a las variables t√©cnicas, que describen los aspectos musicales y ac√∫sticos que se analizan y visualizan en este proyecto. Estas variables son num√©ricas, tanto enteras como flotantes:

- `Danceability`: Qu√© tan adecuada es la canci√≥n para bailar, representada por un valor flotante entre 0.0 y 1.0, donde 0.0 indica m√≠nima aptitud para el baile y 1.0 la m√°xima. Este valor se calcula considerando el tempo, la estabilidad del ritmo, la fuerza del pulso y la regularidad r√≠tmica de la canci√≥n.

- `Energy`: Mide la intensidad y nivel de actividad percibidos de la canci√≥n, con valores entre 0.0 y 1.0. Las canciones con alta energ√≠a suelen ser r√°pidas, fuertes y ruidosas. Para su c√°lculo se consideran el rango din√°mico, el volumen percibido, el timbre, la tasa de inicio de sonidos y la entrop√≠a general.

- `Key`: Describe la tonalidad musical de la canci√≥n mediante una codificaci√≥n num√©rica basada en la notaci√≥n est√°ndar de clases de tono. Si no se logra identificar la tonalidad, se asigna el valor -1.

- `Loudness`: Volumen promedio de la canci√≥n medido en decibelios (dB), calculado a lo largo de toda la pista, usualmente dentro del rango de -60 dB a 0 dB.

- `Speechiness`: Detecta la presencia de palabras habladas en la canci√≥n. Valores altos indican contenido mayormente hablado, mientras que valores bajos corresponden principalmente a m√∫sica instrumental o cantada.

- `Acousticness`: Mide el grado de certeza de que la canci√≥n sea de estilo ac√∫stico, con valores entre 0.0 y 1.0.

- `Instrumentalness`: Indica la probabilidad de que la canci√≥n no contenga voces. Valores cercanos a 1.0 sugieren que la canci√≥n es predominantemente instrumental.

- `Liveness`: Detecta la presencia de p√∫blico en la grabaci√≥n, lo que sugiere que la canci√≥n fue interpretada en vivo. Valores superiores a 0.8 indican una alta probabilidad de actuaci√≥n en directo.

- `Valence`: Mide el car√°cter emocional positivo de la canci√≥n, con valores entre 0.0 y 1.0.

- `Tempo`: Tempo estimado de la canci√≥n, medido en pulsos por minuto (BPM), calculado como el promedio de la duraci√≥n de cada pulso.

- `Duration_ms`: Duraci√≥n total de la canci√≥n expresada en milisegundos.

Por √∫ltimo, la categor√≠a de variables de popularidad agrupa aquellas que reflejan el nivel de consumo y respuesta del p√∫blico, expresadas mediante valores num√©ricos:

- `Stream`: Cantidad de reproducciones en Spotify

- `Views`: N√∫mero de vistas en YouTube

- `Comments`: Cantidad de comentarios en YouTube
